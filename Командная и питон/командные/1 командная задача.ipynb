В данной задаче вам предстоит написать программу определяющую пересекаются ли графики, полученные путём усреднения пиков на чётных и нечётных отрезках сигнала.

Сигнал представлен последовательностью разделенных запятой целых чисел, лежащих в диапазное от 0 до 255. Его можно разбирть на отрезки длиной 256 точек, каждый из которых содержит один пик случайной ширины и высоты расположенный на середине отрезка. Параметры случайного распределения высоты и ширины пиков различаются для чётных и нечётных участков сигнала.

На верхнем левом рисунке построен график данных из Теста №1, на правом верхнем - из Теста №2. На нижних рисунках оранжевым цветом построены усреднённые графики для второго и третьего пиков из тестовых данных, синим цветом - первого и третьего:



Если графики усреднённых чётных и усреднённых нечётных пиков пересекаются, ваша программа должна напечатать 1, если графики не пересекаются - 0. Ваши решения будут тестироваться на более длительных записях сигнала.

```python
# Решение 1 задачи 5 блока 2 этапа трека Нейротехнологии и когнитивные науки
# Олимпиады НТИ 19/20

import numpy as np

data = np.array([int(i) for i in input().split(',')], dtype=np.int)

# В переменные a и b записываем массивы усреднённых значений, соответственно, нечётных и чётных пиков.
# В переменную c записываем поэлементную разность массивов a и b. Очевидно, что i-й элемент массива c 
# больше нуля только когда i-й элемент массива a больше i-того элемента массива b.

a = np.average([data[i:i+256] for i in range(0, len(data), 512)], axis=0)
b = np.average([data[i:i+256] for i in range(256, len(data), 512)], axis=0)
c = a - b

# Если x = np.array([-1, 5, -5, 8], а y = x[x > 0], то y == [5, 8]. Таким образом, в переменную
# len_mult мы записываем произведение числа элементов массива c больших 0 на число элементов массива
# c меньших 0. Очевидно, что произведение равно нулю только когда графики массивов a и b не
# пересекаются, т.е. когда в массиве c - разности массивов a и b нет либо ни одного элемента
# больше нуля, либо ни одного элемента меньше нуля. Отсюда выводим ответ. 
len_mult = len(c[c<0]) * len(c[c>0])

print(1 if len_mult else 0)

```
