```python
import pandas as pd
import numpy as np
import pylab as pl
from numpy import array, arange, abs as np_abs
from numpy.fft import rfft, rfftfreq, fft, ifft
from numpy.random import uniform
from math import sin, pi
import matplotlib.pyplot as plt
import seaborn as sns

from scipy import signal
```

```python
sns.set(rc={'figure.figsize': (12,8)})
```

Read signals file for alpha detection (7 chanels)

```python
df_alpha = pd.read_csv("/home/alex/hack_neuro/записанные сигналы/alpha.dat", sep="\t", 
                 names=['ch0', 'ch1', 'ch2', 'ch3', 'ch4', 'ch5', 'ch6', 'ch7', 'ch8', 'ch9']) \
                .reset_index()
```

```python
df_alpha = df_alpha.drop(['ch7', 'ch8', 'ch9', 'index'], axis=1) # drop fake channels
```

```python
df_alpha
```

```python
df_alpha.ch5.plot()
```

```python
rate = 250.0 # frequency discretisation
t = np.arange(0, 2000, 1/rate) # first 2000 point
x = df_alpha.ch5[:2000].copy()
p = np.abs(np.fft.rfft(x))
f = np.linspace(0, rate/2, len(p))
pl.plot(f[1:], p[1:])
# no alpha on this spectre
```

```python
rate = 250.0
t = np.arange(0, 2000, 1/rate) # [2000:4000] point
x = df_alpha.ch5[2000:4000].copy()
p = np.abs(np.fft.rfft(x))
f = np.linspace(0, rate/2, len(p))
pl.plot(f[1:], p[1:])
# alpha on this spectre, 10-16 Hz
```
