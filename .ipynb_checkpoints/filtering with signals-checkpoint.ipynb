```python
b, a = signal.iirnotch(52, 20, 250)
smooth_data = signal.lfilter(b, a, df_alpha.ch5.copy())
b, a = signal.iirnotch(52.5, 20, 250)
smooth_data = signal.lfilter(b, a, smooth_data)
b, a = signal.iirnotch(53, 20, 250)
smooth_data = signal.lfilter(b, a, smooth_data)
b, a = signal.iirnotch(53.5, 20, 250)
smooth_data = signal.lfilter(b, a, smooth_data)
b, a = signal.iirnotch(118, 20, 250)
smooth_data = signal.lfilter(b, a, smooth_data)
b, a = signal.iirnotch(118.5, 20, 250)
smooth_data = signal.lfilter(b, a, smooth_data)
b, a = signal.iirnotch(65, 20, 250)
smooth_data = signal.lfilter(b, a, smooth_data)
b, a = signal.iirnotch(65.5, 20, 250)
smooth_data = signal.lfilter(b, a, smooth_data)
b, a = signal.iirnotch(66, 20, 250)
smooth_data = signal.lfilter(b, a, smooth_data)
```

```python
rate = 250.0
x = smooth_data
p = np.abs(np.fft.rfft(x))
f = np.linspace(0, rate/2, len(p))
pl.plot(f[1:], p[1:])
```

```python
fs = 250  # Sampling frequency
# Generate the time vector properly
t = np.arange(4505) / fs
data_filt = df_alpha.copy()
for signal_i in df_alpha.columns:
    signala = df_alpha[signal_i].copy() # with frequency of 250
    nqc = fs / 2 # Normalize the frequency
    low = 1 / nqc
    high = 52 / nqc
    b, a = signal.butter(2, [low, high], btype='band')
    data_filt[signal_i] = signal.lfilter(b, a, signala)
```

```python
rate = 250.0
x = data_filt.ch5.copy()
p = np.abs(np.fft.rfft(x))
f = np.linspace(0, rate/2, len(p))
pl.plot(f[1:], p[1:])
```
